```python
def mean_tau(Gfit_coeffs):
    mean = []

    for tw in range(11):
        
        a = list(Gfit_coeffs[tw][0])
        #just truncating the timescales in situ
#         new = [x for x in a if x <= 9e5]
#         new = [x for x in new if x >= 1]
        
        mean.append(np.mean(a))
        
    return mean
```

```python
def std_tau(Gfit_coeffs):
    mean = []

    for tw in range(11):
        
        a = list(Gfit_coeffs[tw][0])
        #just truncating the timescales in situ
        new = [x for x in a if x <= 9e5]
        new = [x for x in new if x >= 1]
        
        mean.append(np.std(new))
        
    return mean
```
