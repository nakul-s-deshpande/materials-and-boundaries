```python
def plot_histogram(Gfit_coeffs,axes,bins):
    
    kwargs = dict(histtype='stepfilled',density = True, alpha=1, edgecolor='k',  bins=bins)
    color = iter(cm.RdYlBu(np.linspace(0, 1, 11)))
    
    ax_hist = axes

    for i in range(10):
        a = list(Gfit_coeffs[i][0])
        clipped_tau = [x for x in a if x <= 9e5]

        c = next(color)
        ax_hist.hist(clipped_tau,facecolor = c,**kwargs,label=int(labels[i]))
        
#         ax_hist.set_ylabel('frequency')
        ax_hist.set_xlabel(r'$\tau_e$ [s]')
        ax_hist.set_xscale('log')
        # ax1.set_yscale('log')
        ax_hist.set_xlim([1,1e4])
        
    return ax_hist
```

```python
def calculate_cdf(Gfit_coeffs):
    
    # bins = np.arange(0, 1e5, 100)
    bins = np.logspace(np.log10(1),np.log10(9e5), 100)

    cdfs = [[] for _ in range(Gfit_coeffs.shape[0])]
    cdf_bins = [[] for _ in range(Gfit_coeffs.shape[0])]
    tau = [[] for _ in range(Gfit_coeffs.shape[0])]

    for i in range(Gfit_coeffs.shape[0]):
        tau[i] = np.asarray(Gfit_coeffs[i])
        cdfs[i], cdf_bins[i] = exposure.cumulative_distribution(tau[i], bins)
#         hists[i], hist_bins[i] = exposure.histogram(tau[i], bins)

    return cdfs, cdf_bins
```

```python
FlatSmooth_tau_cdfs,FlatSmooth_tau_cdf_bins = calculate_cdf(FlatSmooth_Gfit_coeffs)

FlatRough_tau_cdfs,FlatRough_tau_cdf_bins = calculate_cdf(FlatRough_Gfit_coeffs)
```

```python
def plot_cdf(tau_cdf,tau_cdf_bins,axes,bins):

    color = iter(cm.viridis(np.linspace(0, 1, 11)))
    ax_cdf = axes

    for i in range(10):        
        c = next(color)
        ax_cdf.plot(tau_cdf_bins[i],1 - tau_cdf[i],color = c,alpha = 1,label=int(labels[i]))        
        ax_cdf.set_ylabel('1 - cdf')
        ax_cdf.set_xlabel(r'$\tau_e$ [s]')
        ax_cdf.set_xscale('log')
        ax_cdf.set_yscale('log')

        # ax1.set_yscale('log')
#         ax_cdf.set_xlim([1,1e6])
        
    return ax_cdf
```

```python
# Display results
fig = plt.figure(figsize=(3, 4))
axes = np.zeros((2, 5), dtype=np.object)

axes[0, 0] = plt.subplot(2, 1, 1)
axes[0, 1] = plt.subplot(2, 1, 2)


bins = np.logspace(np.log10(2), np.log10(9e5), 100)

ax_cdf = plot_cdf(FlatSmooth_tau_cdfs,FlatSmooth_tau_cdf_bins, axes[0, 0], bins=bins)
ax_cdf.set_title('FlatSmooth')

ax_cdf = plot_cdf( FlatRough_tau_cdfs,FlatRough_tau_cdf_bins, axes[0, 1], bins=bins)
ax_cdf.set_title('FlatRough')



box = ax_cdf.get_position()
lgd = ax_cdf.legend()
ax_cdf.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax_cdf.legend(title='start time (s)',
           loc='center left',
           frameon=False,
           bbox_to_anchor=(1, 0.5),
           title_fontsize=12)

plt.tight_layout()

# plt.savefig('/home/nakul/Documents/Python Scripts/Materials/ensemble_cdfs.png', bbox_inches='tight')
```
